# Feature Engineering
### to select and transfrom relevant features

### Data Transformation for Binary Classification:
-  BENIGN = 0
-  Others = 1

```python
# Mapping of BENIGN to 0, and others to 1
label_binary = ddos_data['Label'].map(lambda a: 0 if a == 'BENIGN' else 1)
ddos_data['Label_Binary'] = label_binary

print(ddos_data['Label_Binary'].value_counts())
ddos_data.head()
```

```python
# Show Binary Classification Labels
bin_labels_count = ddos_data['Label_Binary'].value_counts()
bin_labels_count.index = ['DDoS Attacks', 'BENIGN']
print(bin_labels_count)

plt.figure(figsize=(4, 4))
sns.countplot(data=ddos_data, x='Label_Binary', palette=['#3861f5', '#ff1a1a'])
plt.xticks([0, 1], ['BENIGN', 'DDoS Attacks'])
plt.show()
```

### Data Transformation for Multi-class Classification:
-  BENIGN = 0
- MSSQL = 1
- UDP = 2
- Syn = 3
- LDAP = 4
- Portmap = 5
- NetBIOS = 6

```python
# Mapping of all labels to unique numbers

def map_label(label):
    if label == 'MSSQL':
        label_type = 1
    elif label == 'UDP':
        label_type = 2
    elif label == 'Syn':
        label_type = 3
    elif label == 'LDAP':
        label_type = 4
    elif label == 'Portmap':
        label_type = 5
    elif label == 'NetBIOS':
        label_type = 6
    else:
        label_type = 0
        
    return label_type

ddos_data['Label_Multiple'] = ddos_data['Label'].apply(map_label)

print(ddos_data['Label_Multiple'].value_counts())
```

```python
# Show Multi-class Classification Labels
print(ddos_data['Label'].value_counts())

plt.figure(figsize=(7, 3.5))
sns.countplot(data=ddos_data, x='Label')
plt.show()
```

### Encoding Categorical Features

```python
# Categorical features to dummy variables
non_numeric_features = ['Source_IP', 'Destination_IP']
ddos_dummies = pd.get_dummies(ddos_data[non_numeric_features])

# Drop target and categorical columns to get only features
numeric_features = list(ddos_data.columns.drop(['Source_IP', 'Destination_IP', 'Protocol_Name', 
                                                   'Label', 'Label_Binary', 'Label_Multiple']))

# Combine relevant features to get training data
training_set = ddos_dummies.join(ddos_data[numeric_features])
training_set.columns
```

# Model Fitting, Prediction & Evaluation 

```python
# Set seed for reproducibility
SEED = 21

# the models
models = [RandomForestClassifier(random_state=SEED), LogisticRegression(random_state=SEED, max_iter=300),
          KNeighborsClassifier(), LinearSVC()]
```

## Binary Classification

```python
bin_y = ddos_data['Label_Binary'].values

# Split data into 70% training set and 30% test set
# X: features; y: targets
bin_X_train, bin_X_test, bin_y_train, bin_y_test = train_test_split(training_set, bin_y, test_size=0.3, random_state=SEED)
```

```python
# Lists for evaluation metrics
bin_model = []
bin_accuracy = []
bin_precision = []
bin_recall = [] # Recall = Sensitivity = TPR
bin_f1 =[]
bin_specificity = [] # Specificity = TNR
bin_roc_auc =[]

for model in models:
    model_name = model.__class__.__name__
    model.fit(bin_X_train, bin_y_train) # Training
    model_pred = model.predict(bin_X_test) # Prediction

    cm = metrics.confusion_matrix(bin_y_test, model_pred, labels=model.classes_)
    cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)

    # Mapping of scorer name to scorer function
    accuracy = np.round(accuracy_score(bin_y_test, model_pred), 3)
    precision = np.round(metrics.precision_score(bin_y_test, model_pred), 3)
    recall = np.round(metrics.recall_score(bin_y_test, model_pred), 3)
    f1_score = np.round(metrics.f1_score(bin_y_test, model_pred), 3)
    specificity = np.round(cm[0,0]/(cm[0,0]+cm[0,1]), 3)
    roc_auc = np.round(roc_auc_score(bin_y_test, model_pred), 3)

    # Append results to corresponding lists
    bin_model.append(model_name)
    bin_accuracy.append(accuracy)
    bin_precision.append(precision)
    bin_recall.append(recall)
    bin_f1.append(f1_score)
    bin_specificity.append(specificity)
    bin_roc_auc.append(roc_auc)

    print("\nThe Metrics for {} are: ".format(model_name))

    print('\nAccuracy: {}'.format(accuracy))
    print('Precision: {}'.format(precision))
    print('Recall: {}'.format(recall))
    print('F1_Score: {}'.format(f1_score))
    print('Specificity: {}'.format(specificity))
    print('ROC_AUC: {} \n\n'.format(roc_auc))

    print(classification_report(bin_y_test, model_pred))
    print(cm)
    cm_display.plot()
    plt.show()
    
    fig, ax = plt.subplots()
    ax.plot([0, 1], [0, 1], 'r--')
    RocCurveDisplay.from_predictions(bin_y_test, model_pred, name=str(model_name), ax=ax)
    plt.show()

    print("================================================================================================================")
```

### CNN Model

```python
print(bin_X_train.shape)
print(bin_X_test.shape, '\n')

# Reshape data for CNN Model
train_X = np.expand_dims(bin_X_train, axis=-1)
test_X = np.expand_dims(bin_X_test, axis=-1)
print(train_X.shape)
print(test_X.shape, '\n')

# Convert class labels to one-hot encoding vector
train_Y = to_categorical(bin_y_train, 2)            # num_classes=2
print(train_Y.shape)
```

```python
# Create model
# Add convolution layer
# Stack up layers by adding the desired layers one by one

model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='sigmoid', padding='same', input_shape=(1593, 1)))
model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model.add(Dropout(0.5))
model.add(MaxPooling1D(2))
model.add(Flatten())
model.add(Dense(2, activation='softmax'))

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_X, train_Y, epochs=20, batch_size=32, verbose=0)
```

```python
# Predictions
model_pred = model.predict(test_X)

# Reshape output
cnn_pred = np.argmax(np.round(model_pred), axis=1)

print(bin_y_test.shape, cnn_pred.shape)
```

```python
# Display model architecture
model.summary() 
```

```python
# Show results
model_name = 'ConvolutionalNeuralNetwork'

cm = metrics.confusion_matrix(bin_y_test, cnn_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm)

# Mapping of scorer name to scorer function
accuracy = np.round(accuracy_score(bin_y_test, cnn_pred), 3)
precision = np.round(metrics.precision_score(bin_y_test, cnn_pred), 3)
recall = np.round(metrics.recall_score(bin_y_test, cnn_pred), 3)
f1_score = np.round(metrics.f1_score(bin_y_test, cnn_pred), 3)
specificity = np.round(cm[0,0]/(cm[0,0]+cm[0,1]), 3)
roc_auc = np.round(roc_auc_score(bin_y_test, cnn_pred), 3)

# Append results to corresponding lists
bin_model.append(model_name)
bin_accuracy.append(accuracy)
bin_precision.append(precision)
bin_recall.append(recall)
bin_f1.append(f1_score)
bin_specificity.append(specificity)
bin_roc_auc.append(roc_auc)

print("\nThe Metrics for Convolutional Neural Network are: ")

print('\nAccuracy: {}'.format(accuracy))
print('Precision: {}'.format(precision))
print('Recall: {}'.format(recall))
print('F1_Score: {}'.format(f1_score))
print('Specificity: {}'.format(specificity))
print('ROC_AUC: {} \n\n'.format(roc_auc))

print(classification_report(bin_y_test, cnn_pred))
print(cm)
cm_display.plot()
plt.show()
    
fig, ax = plt.subplots()
ax.plot([0, 1], [0, 1], 'r--')
RocCurveDisplay.from_predictions(bin_y_test, cnn_pred, name='ConvolutionalNeuralNetwork', ax=ax)
plt.show()
print("================================================================================================================")
```

## Multi-class Classification

```python
multi_y = ddos_data['Label_Multiple'].values

# Split data into 70% training set and 30% test set
# X: features; y: targets
multi_X_train, multi_X_test, multi_y_train, multi_y_test = train_test_split(training_set, multi_y, test_size=0.3, 
                                                                            random_state=SEED)
```

```python
# Lists for evaluation metrics
multi_model = []
multi_accuracy = []

for model in models:
    model_name = model.__class__.__name__
    model.fit(multi_X_train, multi_y_train)
    model_pred = model.predict(multi_X_test)

    cm = metrics.confusion_matrix(multi_y_test, model_pred, labels=model.classes_)
    cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)

    # Mapping of scorer name to scorer function
    accuracy = np.round(accuracy_score(multi_y_test, model_pred), 3)

    # Append results to corresponding lists
    multi_model.append(model_name)
    multi_accuracy.append(accuracy)

    print("\nThe Metrics for {} are: ".format(model_name))

    print('\nAccuracy: {}\n'.format(accuracy))

    print(classification_report(multi_y_test, model_pred))
    print(cm)
    cm_display.plot()
    plt.show()

    print("==========================================================================================================")
```

### CNN Model

```python
print(multi_X_train.shape)
print(multi_X_test.shape, '\n')

# Reshape data for CNN Model
train_X = np.expand_dims(multi_X_train, axis=-1)
test_X = np.expand_dims(multi_X_test, axis=-1)
print(train_X.shape)
print(test_X.shape, '\n')

# Convert class labels to one-hot encoding vector
train_Y = to_categorical(multi_y_train, 7)            # num_classes=7
print(train_Y.shape)
```

```python
# Create model
# Add convolution layer
# Stack up layers by adding the desired layers one by one

model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='sigmoid', padding='same', input_shape=(1593, 1)))
model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
model.add(Dropout(0.5))
model.add(MaxPooling1D(2))
model.add(Flatten())
model.add(Dense(7, activation='softmax'))

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
